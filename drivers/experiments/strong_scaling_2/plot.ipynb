```julia
function parseTiming(fname::String)::Real
    f = open(fname)
    for line in eachline(f)
        occursin("Braid", line) && continue
        occursin("wall time", line) && return parse(Float64, split(line)[end])
    end
end
```

```julia
ncores = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024]
tΔ = zeros(length(ncores)) # preallocate mutable types
tθ = zeros(length(ncores))
tΔfmg = zeros(length(ncores))

for i in 1:length(ncores)
    tΔ[i] = parseTiming("lorenz_theta_Delta_nc$(ncores[i])_ml4")
    tΔfmg[i] = parseTiming("lorenz_theta_Delta_fmg_nc$(ncores[i])_ml4")
    tθ[i] = parseTiming("lorenz_theta_nc$(ncores[i])_ml5")
end
```

```julia
pl = plot(ncores, [tΔ tΔfmg tθ],
          linewidth=3,
          dpi=100,
          label=["θ, Δ" "θ, Δ, fmg" "θ"],
          xaxis=("num cores", :log),
          yaxis=("wall time", :log),
)
plot!(pl, ncores, x->parseTiming("lorenz_theta_ml1"), linewidth=3, label="serial")
savefig(pl, "scaling.png")
plot!()
```

```julia
pl = plot(ncores, [tΔ/9 tΔfmg/6 tθ/12],
          linewidth=3,
          dpi=100,
          label=["θ, Δ" "θ, Δ, fmg" "θ"],
          xaxis=("num cores", :log),
          yaxis=("wall time", :log),
)
plot!(pl, ncores, x->parseTiming("lorenz_theta_ml1"), linewidth=3, label="serial")
savefig(pl, "scaling_normalized.png")
plot!()
```
